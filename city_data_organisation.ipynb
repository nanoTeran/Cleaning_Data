```python
cundi = 'cundi_municipios.geojson'
cu = gpd.read_file(cundi)
cu = cu.rename(columns={"munnombre":"nombre","munarea":"area", "muncodigo":"codigo" })

upz = 'upz_bogota.geojson'
up = gpd.read_file(upz)
up = up.rename(columns={"uplarea":"area", "uplcodigo":"codigo", "uplnombre":"nombre"})
up['area'] = ( up['area'] / 1e6 ).round(2)
del up['uplaadmini']
del up['upltipo']

bog_cund = pd.concat([up, cu])
bog_cund.index = bog_cund['codigo']

equipos = pd.read_excel('prueba.xlsx', index_col=0)
rc = gpd.GeoDataFrame(equipos, geometry = gpd.points_from_xy(equipos.y, equipos.x))

with open('bocu.geojson', 'w', encoding='utf-8') as f:
    geojson.dump(bog_cund, f)
```

```python
#Procesar la base maestra para extraer la información de ubicación necesaria
df = pd.read_excel('Base_Maestra.xlsx')
df = df[["TIPO","EQUIPO","DIRECCION", "COORDENADA X", "COORDENADA Y", "MARCA", "MODELO"]]
df = df.rename(columns={"COORDENADA X":"COOR_X", "COORDENADA Y":"COOR_Y"})
gdf = gpd.GeoDataFrame(df, geometry = gpd.points_from_xy(df.COOR_X, df.COOR_Y))
gdf

with open('base_maestra.geojson', 'w', encoding='utf-8') as f:
    geojson.dump(gdf, f, ensure_ascii=False)

```

```python
#Para graficar con la libreria folium
bog_cund = gpd.read_file('bocu.geojson')
map = folium.Map(location=[45, -74], zoom_start=8, tiles="Stamen Terrain")

equipos2 = pd.read_excel('prueba.xlsx')
rc2 = gpd.GeoDataFrame(equipos2, geometry = gpd.points_from_xy(equipos2.y, equipos2.x))

#codigo para permitir que sea leido por geojson
rc2 = rc2.set_crs("EPSG:4326")
rc2 = rc2.set_crs(epsg=4326)

bog_cund = gpd.read_file('bocu.geojson')
map = folium.Map(location=[4.5, -74], tiles="Cartodb Positron", zoom_start=6)

def style_function(feature):
    return {
        "fillOpacity": 0.1,
        "weight": 1,
    }

folium.GeoJson(bog_cund, style_function=style_function,
               popup = folium.GeoJsonPopup(fields=["nombre"])
              ).add_to(map)


folium.GeoJson(
    rc2,
    marker = folium.Circle(radius=4, fill_color="green", fill_opacity=0.4, color="green", weight=1),
    popup = folium.GeoJsonPopup(fields=["Equipo","x", "y"])
    ).add_to(map)
map

map.save('mapa.html')
```
